```python
import datetime
a = df[2][1:].as_matrix()[0]
print(datetime.timedelta(hours=a.hour, minutes=a.minute, seconds=a.second).seconds)
def to_seconds(a):
    return datetime.timedelta(hours=a.hour, minutes=a.minute, seconds=a.second).seconds
d = df[2][1:].as_matrix()
a = np.zeros_like(d)
for i,val  in enumerate(d):
    a[i] = to_seconds(val)
time = df[3][1:].as_matrix()
```

```python
plt.figure(figsize=(16,10))
plt.scatter(a,time)
plt.show()
```

```python
data_x = a[0:20000].reshape(-1,1)/300
data_y = time.reshape(-1,1)
pred = np.linspace(0,max(data_x),1000).reshape(-1,1)
gamma = None
R = rbf_kernel(data_x, pred,gamma = gamma)
C = rbf_kernel(data_x, gamma=gamma)
print(C)
C_star = rbf_kernel(pred, gamma=gamma)

mu_star = np.dot(np.dot(R.T,inv(C)),data_y)
sigma_star = np.dot(np.dot(R.T,inv(C)),R)
```
